```python
import os, glob
import numpy as np
from tqdm import tqdm

DATA_PATH_IMG = "/Users/apple/Desktop/BRAINIAC/data/processed/images"
DATA_PATH_MASK = "/Users/apple/Desktop/BRAINIAC/data/processed/masks"
SLICE_PATH = "/Users/apple/Desktop/BRAINIAC/slices"

os.makedirs(SLICE_PATH, exist_ok=True)
```

```python
import glob

X_files = sorted(glob.glob(f"{DATA_PATH_IMG}/X_*.npy"))
Y_files = sorted(glob.glob(f"{DATA_PATH_MASK}/Y_*.npy"))

count = 0

for xf, yf in tqdm(zip(X_files, Y_files), total=len(X_files)):

    X = np.load(xf, mmap_mode="r")
    Y = np.load(yf, mmap_mode="r")

    for z in range(X.shape[2]):

        if np.sum(Y[:,:,z]) > 0:   

            img = X[:,:,z,:] 
            mask = Y[:,:,z]

            np.save(f"{SLICE_PATH}/img_{count}.npy", img.astype(np.float32))
            np.save(f"{SLICE_PATH}/mask_{count}.npy", mask.astype(np.float32))

            count += 1

print("Saved slices:", count)
```
