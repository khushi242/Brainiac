```python
import os
import shutil
import glob

BASE = "/Users/apple/Desktop/BRAINIAC"

raw_dir = f"{BASE}/data/raw"
proc_img = f"{BASE}/data/processed/images"
proc_mask = f"{BASE}/data/processed/masks"

os.makedirs(raw_dir, exist_ok=True)
os.makedirs(proc_img, exist_ok=True)
os.makedirs(proc_mask, exist_ok=True)

print("Folders created")
```

```python
old_proc = "/Users/apple/Desktop/BRAINIAC/processed"

for f in glob.glob(f"{old_proc}/X_*.npy"):
    shutil.move(f, f"{proc_img}/{os.path.basename(f)}")

for f in glob.glob(f"{old_proc}/Y_*.npy"):
    shutil.move(f, f"{proc_mask}/{os.path.basename(f)}")

print("Moved processed files")
```
