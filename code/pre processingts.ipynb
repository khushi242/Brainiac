```python
import os
import cv2
import numpy as np
from tqdm import tqdm

# рџ”№ INPUT FOLDER (tumor-less images)
raw_no_tumor = "/Users/apple/Desktop/BRAINIAC/data/raw/no_tumor"

# рџ”№ OUTPUT FOLDERS (already used by your training)
processed_img = "/Users/apple/Desktop/BRAINIAC/data/processed/images"
processed_mask = "/Users/apple/Desktop/BRAINIAC/data/processed/masks"

# create folders if missing
os.makedirs(processed_img, exist_ok=True)
os.makedirs(processed_mask, exist_ok=True)

print("Checking folder...")
print("Exists:", os.path.isdir(raw_no_tumor))
print("Files:", os.listdir(raw_no_tumor))

files = os.listdir(raw_no_tumor)

for f in tqdm(files):
    
    img_path = os.path.join(raw_no_tumor, f)

    # skip non-image files
    if not f.lower().endswith(('.png','.jpg','.jpeg','.bmp','.tiff')):
        continue

    img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)

    if img is None:
        continue

    h, w = img.shape

    # create EMPTY mask
    empty_mask = np.zeros((h, w), dtype=np.uint8)

    # new name so it doesn't clash
    name = os.path.splitext(f)[0] + "_no_tumor.png"

    # save image + mask
    cv2.imwrite(os.path.join(processed_img, name), img)
    cv2.imwrite(os.path.join(processed_mask, name), empty_mask)

print("вњ… Tumor-less data added successfully!")
```
